```python
import numpy as np
```

```python
def get_matrix_minor(A, lin, col):
    row = np.concatenate((A[:lin], A[lin + 1:]))
    minor = np.concatenate((row[:, :col], row[:, col + 1:]), axis=1)
    
    return minor
```

```python
def get_matrix_deternminant(A):
    n = len(A)
    
    # base case for 2x2 matrix
    if n == 2:
        
        return A[0][0] * A[1][1] - A[0][1] * A[1][0]

    determinant = 0
    for col in range(n):
        cof = ((-1) ** col) * get_matrix_deternminant(get_matrix_minor(A, 0, col))
        determinant += cof * A[0][col]
        
    return determinant
```
